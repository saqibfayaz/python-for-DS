## Adding Match Summary

```python
import pandas as pd
import json
```

```python
with open('/Users/saqib/Downloads/t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)
data[0]['matchSummary']
```

```python
#convert json file to pd.DataFrame

df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
# rename scorcard as match_id so that later on,
# i can use it as a key to join different tables

df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace= True)
df_match.head()
```

```python
match_ids_dict = {}

for index,row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']

match_ids_dict
```

```python
df_match.to_csv('/Users/saqib/Downloads/dim_match_summary.csv', index = False)
```

## 2: Adding Batting Summary

```python
# extebding array eg eg

a = [1,2]
b = [3,4]
a.extend(b)
a
```

```python
with open('/Users/saqib/Downloads/t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    all_records = []
    
    for rec in data:
        all_records.extend(rec['battingSummary'])
all_records
```

```python
# now create df from all_records
df_batting = pd.DataFrame(all_records)
df_batting.head(11)
```

```python
# let's convert dismissal column into out/notout form
# to make it uniform and helpful for analysis
# you can see the filled row in that particular column is out and empty one is not out

df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: "out" if len(x) > 0 else "not_out" )
df_batting
```

```python
# now delete/drop dismissal column as it's unusable

df_batting.drop(columns=['dismissal'], inplace= True)
df_batting.head(20)
```

```python
# remove extra characters in rows of 'batsmanName' column.
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('†','')) 
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0','')) 
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
```

```python
df_batting.head(10)
```

```python
df_match.head()
```

```python
# after looking at these two datasets we need to somehow connect them,
# the only way is to somehow join df_match('team1' , 'team2')
# with match column of df_batting
# we can create a match_ids_dict in df_match to join team1,team2
# in both combinations like 'team1 vs team 2' or 'team2 vs team1' to avoid mismatches
'''
match_ids_dict = {}

for index,row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']

match_ids_dict

'''
```

```python
match_ids_dict['Namibia Vs Sri Lanka']
```

```python
# now add new column 'match_id' to df_batting by using
# values of'match_ids_dict' as the rows of new column 'match_id'

df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
df_batting.head(20)
```
