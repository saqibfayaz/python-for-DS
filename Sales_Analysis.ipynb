```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime 
```

```python
df = pd.read_excel(r'/Users/saqib/Downloads/superstore_sales.xlsx')
```

```python
product_name_sales = pd.DataFrame(df.groupby(['product_name']).sum(numeric_only= True)['sales'])[:5] # we could do this instead of using .head(5)
product_name_sales.sort_values(by = ['sales'], ascending= False, inplace= True)
product_name_sales
```

```python
product_name_sales.index
```

```python
plt.figure(figsize=(5,3)) # to adjust plot size
plt.bar(product_name_sales.index, product_name_sales['sales'])
plt.xticks(rotation = 'vertical', size = 8) # to rotate labels on x-axis
plt.show()
```

```python
df[['segment', 'profit']]
```

```python
product_name_profit = pd.DataFrame(df.groupby(['segment']).sum(numeric_only= True)['profit'])[:5] # we could do this instead of using .head(5)
product_name_profit.sort_values(by = ['profit'], ascending= False, inplace= True)
product_name_profit
```

```python
sns.barplot(x = 'segment', y = 'profit', data = product_name_profit)
```

```python
# Could not interpret input 'segment'
# so let's reset the index to make 'segment a column'
sns.barplot(x = 'segment', y = 'profit', data = product_name_profit.reset_index())
plt.show()
```

```python
df.head(1)
#df.dtypes, so it's datetime , we need to convert 
#it before extracting month
```

```python
month = df['order_date'].astype(str)
month = month.apply(lambda x: datetime.strptime(x, '%Y-%m-%d' ))
df['month'] = month.map(lambda x: x.month)
df
```

```python
date = df.groupby(['year', 'month']).sum()
date
```

```python
sns.barplot(x = 'year', y = 'sales', data = date.reset_index())
```

```python
countries = pd.DataFrame(df.groupby('country').sum(numeric_only= True)['sales'])
countries.sort_values('sales', inplace= True, ascending= False)
countries_top_10 = countries.head(10)
countries_top_10
```

```python
countries_top_10['sales'].plot(kind = 'pie', autopct = '%1.1f%%', figsize = (10,5))
plt.title('Top 10 Countries by Sales')
```

```python
countries_top_10_shipping_cost =df.groupby(['country']).agg({'shipping_cost': 'max'}).sort_values('shipping_cost', ascending = False).head(10)
```

```python
countries_top_10_shipping_cost
```

```python
plt.figure(figsize= (5,3))
plt.bar(countries_top_10_shipping_cost.index, countries_top_10_shipping_cost['shipping_cost'], color = 'red')
plt.xticks(rotation = 'vertical', size = 6)
plt.show()
```
