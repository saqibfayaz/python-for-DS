# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'/Users/saqib/Downloads/Ice Cream Ratings.csv')
df = df.set_index
df
```

```python
df = df.set_index('Date')
df
```

```python
df.plot(kind = 'line')
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
print(plt.style.available)
plt.style.use('fivethirtyeight')
```

```python
df.plot(kind = 'line',title = 'Ice Cream Ratings', xlabel ='Daily Ratings',ylabel = 'scores')
```

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'bar', stacked = True)
```

```python
df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

```python
df.plot.barh(stacked = True)
```

```python
df.plot.scatter(x = 'Texture Rating', y ='Overall Rating', s = 500, c = 'green')
```

```python
df.plot.hist(bins = 10)
#histograms is very important for looking at distribution of variables
```

```python
df.boxplot()
# bottom'_' = minimum value
# bottom of box = 25th percentile
#green line = 50th percentile
#top of box = 75th percentile
# top '_' = maximum value
#in our example in 'flavour rating',we have low minimum(median),.... 
# high maximum and skews towards the lower range
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y ='Flavor Rating', figsize=(10,6))
```
