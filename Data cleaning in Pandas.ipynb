# Data cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r'/Users/saqib/Downloads/Customer Call List.xlsx')
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns= 'Not_Useful_Column')
df
```

```python
''' One Way
df["Last_Name"] = df['Last_Name'].str.lstrip('...')
df["Last_Name"] = df['Last_Name'].str.lstrip('/')
df["Last_Name"] = df['Last_Name'].str.rstrip('_')
'''

#use single statement to remove all at once

df['Last_Name'] = df['Last_Name'].str.strip('123._/')
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
#using regular expressions to replace unwanted characters
df
```

before adding '-' to phone numbers to format it properly.
we need to convert whole column into string because there are some null values in string format already.
if you use lamda function to add '-' to whole column, it will show error. eg like this

```python

df['Phone_Number'].apply(lambda x:x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

so let's change column to string first.we cannot do it simply by this
`str(df["Phone_Number"])` becauseif we did, it will crearte as mess, so we neeed to use lamba function to convert into string

```python
#this will not work the way we need it to
str(df["Phone_Number"])
```

```python
#let's fix it
df['Phone_Number'] = df['Phone_Number'].apply(lambda x:str(x))
```

```python
#now this will work

df['Phone_Number'].apply(lambda x:x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x:x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
#df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
```

```python
df
```

```python
df[["Street_address", "State", 'Zip_Code']] = df['Address'].str.split(',',2, expand = True)
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
```

```python
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
```

```python
df
```

```python
df = df.replace('N/a','')
```

```python
df
```

```python
df = df.fillna('')
```

```python
for x in df.index:
    if df.loc[x,'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace= True)
```
