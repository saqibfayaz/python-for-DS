# Merge, Join , and Concatenate


```python
import pandas as pd
```

```python

df1 = pd.read_csv(r'/Users/saqib/Downloads/LOTR.csv')
df1
```

```python
df2 =pd.read_csv(r'/Users/saqib/Downloads/LOTR 2.csv')
df2
```

```python
df1.merge(df2) # by default it will do inner join operation 
```

```python
df1.merge(df2, how = 'inner') 
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID']) 
#notice first_name has bee shown seperately shows that we are not joining on them
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName']) 
```

```python
df1.merge(df2, how = 'outer') 
```

```python
df1.merge(df2, how = 'left') 
```

```python
df1.merge(df2, how = 'right') 
```

```python
df1.merge(df2, how = 'cross') 
```

```python
df1.join(df2) # it will not work like merge but neeeds some arguments
```

```python
df1.join(df2, on ='FellowshipID', how = 'outer', lsuffix= '_Left', rsuffix= '_Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix='_Left', rsuffix= '_Right', how = 'outer')
df4
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],join= 'outer')
```

```python
pd.concat([df1,df2],join= 'outer', axis = 1)
```

```python
df1.append(df2) #use concat instead because append is gonna be reomved from future version of pandas
```
